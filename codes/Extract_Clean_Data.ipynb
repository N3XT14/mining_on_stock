```python
import numpy as np
import pandas as pd

# assume 'stocks' is a list of 100 dataframes, each with 100 records
window_size = 100
input_data = []
output_data = []

# create sliding windows for each stock
for df in train_data:
    for i in range(len(df) - window_size):
        window = df[i:i+window_size]
        input_data.append(window.values)
        next_price = df.iloc[i+window_size]['Label']
        if next_price == 'decrease':
            output_data.append([1, 0, 0]) 
        elif next_price == 'no big change':
            output_data.append([0, 1, 0])  
        else:
            output_data.append([0, 0, 1])  

input_data = np.array(input_data)
output_data = np.array(output_data)
```
